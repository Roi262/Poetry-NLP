We start by importing pandas and reading the csv.

```python
import pandas as pd

# df = (pd.read_csv("kaggle_poem_dataset.csv", nrows=1_000_000, 
#                   encoding="ISO-8859-1", usecols=['Title', 'Id']))

df = pd.read_csv("Data/kaggle_poem_dataset.csv")
```

Now we want to get two lists, one for the title and the author, and the second for the content of the poem.

```python
authors_and_titles = []

for index, rows in df.iterrows(): 
    # Create list for the current row 
    my_list =[rows.Author, rows.Title] 
      
    # append the list to the final list 
    authors_and_titles.append(my_list) 
    
authors_and_titles

```

```python
poems_content = [_ for _ in df['Content']]

# remove first element from the list (it is creators note)
poems_content.pop(0)

# poems_content
len(poems_content)
```

Now let us see the first poem:

```python
lines_in_poem = poems_content[0].split('\n') 
for i in range(lines_in_poem):
    print("Author: {}, Poem: {}\n {}"
          .format(authors_and_titles[1][0],
                  authors_and_titles[1][1],
                  lines_in_poem[i]))

# lines_in_poem[0]
# lines_in_poem[1]
```
